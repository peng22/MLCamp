```python
mean_price = focus_df['price'].mean()
mean_price
```

```python
focus_df['binary_price']=(focus_df['price']>mean_price).astype(int)
focus_df['binary_price']
```

```python
from sklearn.model_selection import train_test_split
df_full_train, df_test = train_test_split(focus_df,test_size=0.2,random_state=42)
```

```python
df_train,df_val = train_test_split(df_full_train,test_size=0.25,random_state=42)
```

```python
print(len(df_train),len(df_val),len(df_test))
```

```python
y_train = df_train['binary_price']
del df_train['binary_price']
del df_train['price']
y_val = df_val['binary_price']
del df_val['binary_price']
del df_val['price']

y_test = df_test['binary_price']
del df_test['binary_price']
del df_test['price']
```

```python
from sklearn.metrics import mutual_info_score
max_score = -1
max_cor= ''
for col in categorical_columns:
    score = mutual_info_score(df_train[col],y_train)
    print(score,col)
    if score > max_score:
        max_score = score
        max_cor= col


print(f'max corr with col {max_cor} with score {max_score}')
```

```python
df_train.isnull().sum()
df_train=df_train.fillna(0)
df_val=df_val.fillna(0)
df_test=df_test.fillna(0)
```

```python
from sklearn.feature_extraction import DictVectorizer
df_train.isnull().sum()
dv =DictVectorizer(sparse=False)

dicts_train = df_train.to_dict(orient='records')
encoded_df_train=dv.fit_transform(dicts_train)
dicts_val = df_val.to_dict(orient='records')
encoded_df_val=dv.transform(dicts_val)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
model.fit(encoded_df_train,y_train)
y_pred = model.predict(encoded_df_val)
y_pred
```

```python
y_val
```

```python
y_correct = (y_pred == y_val).astype(int)
acc = y_correct.mean()
acc_rounded = acc.round(2)
acc_rounded
```

```python
def test_removing_feature(feature):
    df_train_without = df_train.copy()
    df_val_without = df_val.copy()
    print(df_train_without.columns)
    del df_train_without[feature]
    del df_val_without[feature]

    dv =DictVectorizer(sparse=False)

    dicts_train = df_train_without.to_dict(orient='records')
    encoded_df_train_without=dv.fit_transform(dicts_train)
    dicts_val = df_val_without.to_dict(orient='records')
    encoded_df_val_without=dv.transform(dicts_val)

    model = LogisticRegression(solver='liblinear', C=10, max_iter=1000, random_state=42)
    model.fit(encoded_df_train_without,y_train)
    y_pred = model.predict(encoded_df_val_without)
    
    return {feature : abs(acc - (y_pred == y_val).mean())}
```

```python
for feature in ['year','engine_hp','transmission_type','city_mpg']:
    print(feature,test_removing_feature(feature))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

def train_and_evaluate(X_train, y_train , X_val, y_val,alpha):
   
    model = Ridge(alpha=alpha)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    return rmse

alphas = [0, 0.01, 0.1, 1, 10]
min_rsme = 10
min_alpha =100
for alpha in alphas:
    zero_rmse =train_and_evaluate(encoded_df_train,ylog_train,encoded_df_val,ylog_val,alpha)
    print(alpha,zero_rmse)
    if zero_rmse < min_rsme:
        min_rsme = zero_rmse
        min_alpha = alpha

print(min_alpha,min_rsme) 
```
